```python
import json
import pandas as pd
```

```python
%%time
size = 1000
range_index = 1000

rows = []

for i in range(range_index):
    filename = f'spotify_million_playlist_dataset/data/mpd.slice.{i*size}-{(i+1)*size -1}.json'

    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # data['playlists'] is a list, each element in it is a playlist object
    playlists = data['playlists']

    # We want to merge all the tracks from all playlists into one table
    for playlist in playlists:
        pid = playlist['pid']
        playlist_name = playlist['name']
        tracks = playlist['tracks']
        
        for track in tracks:
            track_uri = track['track_uri']
            # extract real track_id
            # such as spotify:track:0yLjkpa51CN1uiryMDOrtt -> 0yLjkpa51CN1uiryMDOrtt
            track_id = track_uri.split(':')[-1]
            
            rows.append({
                'pid': pid,
                'playlist_name': playlist_name,
                'track_id': track_id,
                'artist_name': track.get('artist_name'),
                'track_name': track.get('track_name')
            })

df_playlists = pd.DataFrame(rows)
df_playlists.head()
```

```python
selected_cols = ['id', 'danceability', 'energy', 'key', 'loudness', 'speechiness', 
                 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 
                 'occurrence_count']
new_df2 = new_df[selected_cols]
new_df2.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Feature columns that need to be normalized (excluding 'id')
feature_cols = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 
                'acousticness', 'instrumentalness', 'liveness', 'valence', 
                'tempo', 'occurrence_count']

df_for_scaling = new_df2.copy()

# Separate the sub-DataFrame to be normalized
features_to_scale = df_for_scaling[feature_cols]

# initial StandardScaler
scaler = StandardScaler()

# Fit and transform
features_scaled = scaler.fit_transform(features_to_scale)

# Convert back to DataFrame and add column names
df_scaled_features = pd.DataFrame(features_scaled, columns=feature_cols)

# If necessary, you can concatenate the 'id' column back together with the normalized features.
df_scaled = pd.concat([df_for_scaling[['id']], df_scaled_features], axis=1)

df_scaled.head()
```

```python
df_scaled.to_csv('scaled_songs.csv', index=False)
```
